```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os 
```

```python
A = imread("Dog.jpg")
X = np.mean(A, -1)
```

```python
img = plt.imshow(256-X)
img.set_cmap("gray")
plt.axis("off")
plt.show()
```

```python
U, S, VT = np.linalg.svd(X, full_matrices = False)
S = np.diag(S)

j = 0
for r in (5, 20, 50, 100):
    Xapprox = U[:, :r] @ S[0:r, :r] @ VT[:r, :]
    plt.figure(j+1)
    j += 1
    img = plt.imshow(256-Xapprox)
    img.set_cmap("gray")
    plt.axis("off")
    plt.title("r = " + str(r))
    plt.show
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title("Singular Values")
plt.show

plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title("Singular Values: Cummulative Sum")
plt.show()
```
